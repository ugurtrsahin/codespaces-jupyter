```
import numpy as np
import matplotlib.pyplot as plt

# Evrensel bir polinom fonksiyonu tanımlıyoruz
def universal_polynomial(x, coefficients):
    """
    Polinom fonksiyonu: f(x) = a_n*x^n + a_(n-1)*x^(n-1) + ... + a_0
    :param x: Girdi değeri
    :param coefficients: Polinom katsayıları
    :return: Polinom değeri
    """
    return sum(c * (x ** i) for i, c in enumerate(coefficients))

# Farklı katsayılarla fonksiyonu simüle ediyoruz
coefficients_list = [
    [1, -1, 1],  # x^2 - x + 1
    [1, 0, -1],  # x^2 - 1
    [0, 1, 0, -1]  # x^3 - 1
]

# x değerlerini oluşturuyoruz
x = np.linspace(-2, 2, 400)

# Grafiği oluşturuyoruz
plt.figure(figsize=(10, 6))

for coefficients in coefficients_list:
    y = universal_polynomial(x, coefficients)
    plt.plot(x, y, label=f'Coefficients: {coefficients}')

plt.title('Evrensel Polinom Fonksiyonu Simülasyonu')
plt.xlabel('x Değeri')
plt.ylabel('f(x) Değeri')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Evrensel bir polinom fonksiyonu tanımlıyoruz
def universal_polynomial(x, coefficients):
    return sum(c * (x ** i) for i, c in enumerate(coefficients))

# Farklı katsayılarla fonksiyonu simüle ediyoruz
coefficients_list = [
    [1, -1, 1],  # x^2 - x + 1
    [1, 0, -1],  # x^2 - 1
    [0, 1, 0, -1],  # x^3 - 1
    [0, 0, 1, -1]   # x^3 - x
]

# x değerlerini oluşturuyoruz
x = np.linspace(-2, 2, 400)

# Koordinat sistemlerini oluşturuyoruz
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Kartezyen Koordinat Sistemi
axs[0, 0].set_title('Kartezyen Koordinat Sistemi')
for coefficients in coefficients_list:
    y = universal_polynomial(x, coefficients)
    axs[0, 0].plot(x, y, label=f'Coefficients: {coefficients}')
axs[0, 0].axhline(0, color='black', lw=0.5, ls='--')
axs[0, 0].axvline(0, color='black', lw=0.5, ls='--')
axs[0, 0].legend()
axs[0, 0].grid()

# Kutupsal Koordinat Sistemi
theta = np.linspace(0, 2 * np.pi, 400)
r1 = universal_polynomial(np.cos(theta), coefficients_list[0])
r2 = universal_polynomial(np.cos(theta), coefficients_list[1])
axs[0, 1].set_title('Kutupsal Koordinat Sistemi')
axs[0, 1].polar(theta, r1, label=f'Coefficients: {coefficients_list[0]}')
axs[0, 1].polar(theta, r2, label=f'Coefficients: {coefficients_list[1]}')
axs[0, 1].legend()

# Aksiyel Koordinat Sistemi
y = np.linspace(-2, 2, 400)
for coefficients in coefficients_list:
    axs[1, 0].plot(y, universal_polynomial(y, coefficients), label=f'Coefficients: {coefficients}')
axs[1, 0].set_title('Aksiyel Koordinat Sistemi')
axs[1, 0].axhline(0, color='black', lw=0.5, ls='--')
axs[1, 0].axvline(0, color='black', lw=0.5, ls='--')
axs[1, 0].legend()
axs[1, 0].grid()

# Üçgen Koordinat Sistemi
triangle_x = np.linspace(-2, 2, 400)
triangle_y = universal_polynomial(triangle_x, coefficients_list[0])
axs[1, 1].set_title('Üçgen Koordinat Sistemi')
axs[1, 1].plot(triangle_x, triangle_y, label=f'Coefficients: {coefficients_list[0]}')
axs[1, 1].fill_between(triangle_x, triangle_y, color='lightgray', alpha=0.5)
axs[1, 1].axhline(0, color='black', lw=0.5, ls='--')
axs[1, 1].axvline(0, color='black', lw=0.5, ls='--')
axs[1, 1].legend()
axs[1, 1].grid()

plt.tight_layout()
plt.show()
```
